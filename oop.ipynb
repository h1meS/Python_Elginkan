```python
class ataSinif():
    def __init__(self,ad,soyad):
        self.ad=ad
        self.soyad=soyad
    
    def yazdir(self):
        print(self.ad,self.soyad)

class cocukSinif(ataSinif):
    def __init__(self,gozRengi,sacRengi,boy):
        self.gozRengi=gozRengi
        self.sacRengi=sacRengi
        self.boy=boy
    
    def yazdir(self):
        print(self.gozRengi,self.sacRengi,self.boy)

cocuk1 = cocukSinif("Sarı","Mavi","Uzun")
cocuk1.yazdir()
```

```python
class Ata():
    def __init__(self,ad,soyad):
        self.ad=ad
        self.soyad=soyad

class Cocuk(Ata):
    def __init__(self,ad,soyad,numara,sinif):
        super().__init__(ad, soyad)
        self.no=numara
        self.sinif=sinif

    def __yazdir(self):
        print(self.ad,self.soyad,self.no,self.sinif)


cocuk1 = Cocuk("Semih", "Sarı",1234,2)
cocuk1.__yazdir()
```

```python
class Araba():
    def Tekerlek(self):
        return ("Arabanın tekerlek sayısı 4")
    def Kontrol(self):
        return ("Arabanın kontrolü direksiyon ile sağlanır")
    def Yakit (self):
        return ("Arabanın yakıtı fosil yakıttır")

class Bisiklet():
    def Tekerlek(self):
        return("Bisikletin tekerlek sayısı 4")
    def Kontrol(self):
        return("Bisikletin kontrolü gidon ile sağlanır")
    def Yakit(self):
        return("Bisikletin gücü insandır")

bisiklet=Bisiklet()
araba=Araba()
for tasit in (bisiklet,araba):  
    print(tasit.Kontrol())
    print(tasit.Tekerlek())
    print(tasit.Yakit())
```

```python
class Maas():
    def __init__(self,yetki,cocuk,yds,kidem):
        self.yetki=yetki
        self.cocuk=cocuk
        self.yds=yds
        self.kidem=kidem

    def maasHesap(self):
        
        if(self.yds<50):
            prim=0
        elif(50<self.yds<70):
            prim=500
        elif(70<self.yds<90):
            prim=1000
        else:
            prim=1500

        liste=(15000,7500,5000)
        return (liste[self.yetki-1] + (self.kidem*500) + (self.cocuk*750) + prim)

    def maasYazdir(self):
        print(f"Maaş: {self.maasHesap()}")

class Kullanicilar(Maas):
    def __init__(self,ID,isim,eposta,telefon,yetki,cocuk,yds,kidem):
        super().__init__(yetki, cocuk, yds, kidem)
        self.__ID=ID
        self.isim=isim
        self.eposta=eposta
        self.telefon=telefon

    def bilgileriYazdir(self):
        print(f"Personel ismi: {self.isim} E-posta: {self.eposta} Telefon: {self.telefon}  ")
 
    def getID(self):
        if (self.yetki==1):
            print(f"ID: {self.__ID}") 
        else:
            print("Yetkiniz Yok!")    

    def setID(self,ID):
        if(self.yetki==1):
            self.__ID=ID
        else:
            print("Yetkiniz Yok!")

class Personel(Kullanicilar):

    def __init__(self,ID,isim,eposta,telefon,yetki,cocuk,yds,kidem):
        super().__init__(ID, isim, eposta, telefon, yetki, cocuk, yds, kidem)


mudur=Personel(1, "Aysel Yıldırım", "ayselyil@gmail.com", "05051234567", 1, 2, 2, 15)
mudur.bilgileriYazdir()
mudur.maasYazdir()        
```
